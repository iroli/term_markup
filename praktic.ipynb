```python
# Open and clean tex doc
def next_tex_doc(address):
    doc = "" # returning value, string
    texdoc = []  # a list of string representing the latex document in python
    # read the .tex file, and modify the lines
    with open(address, 'UTF-8') as texdoc:
        for line in texdoc:
            l = unicode(line, 'utf-8')
            if l.find('%') != 0:
              #  print l
                if len(l) > 0:
                    l = l.lower().replace("~---", "")
                    l = l.replace(".~", ".")
                    l = l.replace("{", ' ').replace("}", ' ')
                    l = l.replace("(", ' ').replace(")", ' ')
                    l = l.replace("[", ' ').replace("]", ' ')
                    l = l.replace("<", ' ').replace(">", ' ')
                    l = l.replace(".", ' ')
                    l = l.replace(",", ' ')
                    l = l.replace("!", ' ')
                    l = l.replace("?", ' ')
                    l = l.replace(";", ' ').replace(":", ' ')
                    doc = doc + l.replace("\n", " ")
    doc = doc + ' '
    return doc
```
