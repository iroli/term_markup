```python
def round_double(term, num, orig_term):
    term = nitem = morph.parse(term)[0]
    nitem = term.normal_form
    c = nitem.count(" ")
    if c == 0:
        norm_library.append([nitem, num, orig_term])
    else:
        n_lst = nitem.split()
        nitem = ""
        for word in n_lst:
            w = morph.parse(word)[0]
            nitem = nitem + ' ' + w.normal_form
        nitem = nitem[1:]
        norm_library.append([nitem, num, orig_term])
    while c > 0:
        d = nitem.find(" ")
        nitem = nitem[d+1:] + ' ' + nitem[:d]
        norm_library.append([nitem, num, orig_term])
        c = c - 1
```

```python
def x_pos(st):
    c = st.count(')')
    f = st.find('(')
    while c > 1:
        f = f + st[f + 1 :].find('(') + 1
        c = c - 1
    return f
```
